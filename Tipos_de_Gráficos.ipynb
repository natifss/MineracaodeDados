# Historiograma

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
idades = np.random.randint(0, 100, size=1000) #distribuição aleatória de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

```python
def Historiograma(Style='ggplot'):
  plt.style.use(Style)
  # Vamos gerar um conjunto de dados aleatórios para as idades
  np.random.seed(0)
  idades = np.random.randint(0, 200, size=2000) #distribuição aleatória de valor inteiro

  # Criar o histograma
  plt.hist(idades, bins=20, edgecolor='black')

  plt.title('Distribuição das Idades')
  plt.xlabel('Idade')
  plt.ylabel('Frequência')

  # Mostrar o gráfico
  plt.show()
     
```

```python
Historiograma('dark_background')
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: Esta distribuição aleatória distribui as idades geradas entre 0 e 100 sem ter nenhum outro tipo de fator.
No caso da distribuição normal, as idades seguirão um padrão e serão acumuladas em um determinado espaço de idade, não aleatóriamente como no exemplo acima.

```python
def GraficodePizza(Style='ggplot'):
  # Dados
  quantidade = [650, 550]
  porcentagemF = round((650 * 100)/1200, 1)
  porcentagemM = round((550 * 100)/1200, 1)
  porcentFstring = str(porcentagemF)
  porcentMstring = str(porcentagemM)
  generos = ['Feminino - ' + porcentFstring + '%','Masculino - ' + porcentMstring + '%']
  plt.style.use(Style) 
  # Criar o gráfico de pizza
  plt.pie(quantidade, labels=generos, startangle=90, colors=['#FFC0CB', '#0000ff'], wedgeprops={'width': 0.4})

  plt.title('Distribuição de Gênero')


  # Mostrar o gráfico
  plt.show()
```

```python
GraficodePizza('dark_background')
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: 
 - A linha mediana tem o valor de aproximadamente 40.
 - O valor máximo é 85, e o valor mínimo é -21.
 - A amplitude do gráfico é 105, portanto, tem uma grande variação de dados.
 - O valor Q1 é aproximadamente 22, e o valor Q3 é aproximadamente 48.
 - Os dados são negativamente assiméticas, pois a mediana está mais próxima ao Q3.
 - Não há dados outliers.

```python
def DiagramadeCaixa(Style):
  plt.style.use(Style)
  # Dados
  distribuicao = np.random.normal(40, 20, 120)

  # Criar o gráfico de caixa
  plt.boxplot(distribuicao)

  plt.title('Distribuição Personalizada')
  plt.xlabel('Dados')
  plt.ylabel('Valores')

  # Mostrar o gráfico
  plt.show()
```

```python
DiagramadeCaixa('dark_background')
```

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R: A equipe A teve pontuações mais baixas, a maioria entre 90 e 70.
A equipe D também teve a maioria das pontuações com os mesmos resultados, entre 100 e 80. Já a equipe C, teve as pontuações dispersas, grande parte entre 90 e 70, porém algumas chegaram à atingir 120 pontos.


```python
def GraficodeViolino (Style='ggplot'):
  plt.style.use(Style)
  # Criar o gráfico de violino
  equipes = df.columns.tolist()
  plt.figure(figsize=(6, 4))
  plt.violinplot(df)

  # Personalizar o gráfico
  plt.title('Distribuição da Pontuação por Equipe')
  plt.xlabel('Equipes')
  plt.ylabel('Pontuação')

  plt.xticks(np.arange(1, len(equipes)+1), equipes)

  # Mostrar o gráfico
  plt.show()
```

```python
GraficodeViolino('bmh')
```
